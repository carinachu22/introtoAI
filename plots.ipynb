# 0 Preliminaries 
First, import numpy, matplotlib and seaborn for plots   


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
%matplotlib inline
```

# 1 Initial Description of Data 
First, we load the csv file and look at the overview of the data

```python
df_penguins=pd.read_csv("penguins.csv")
```

```python
df_penguins.info()
```

We analyse the following things: 
1. The population for each of the 3 species 
2. Species based on the non-numerical features: island and gender
3. Pairplot of bill_length, bill_depth, flipper_length, body_mass

```python
sns.set_style('white')
colors = ['#8db5e6', '#9ed99d', '#ffb6c1']
font_size = 16
ax_species = df_penguins['species'].value_counts().plot(kind='bar',color=colors)

# Add y-axis label
ax_species.set_ylabel('Count')
ax_species.set_xlabel('Species')
# Add graph title
ax_species.set_title('Number of Penguins by Species')

plt.show()
```

```python
gender_colors = ['#8db5e6', '#ffb6c1']
ax_species = sns.catplot(x='species', hue='sex', kind='count', data=df_penguins,palette=gender_colors)

# Set labels and title for each subplot
for ax in ax_species.axes.flat:
    ax.set_ylabel('Count',fontsize=font_size)
    ax.set_xlabel('Species',fontsize=font_size)
    ax.set_title('Number of Penguins by Gender and Species',fontsize=font_size)
plt.show()
```

```python
sns.swarmplot(x=df_penguins.island,y=df_penguins.bill_length_mm,hue=df_penguins.species, palette=colors)

# Add graph title
plt.title('Bill Length by Island and Species',fontsize=font_size)
plt.xlabel("Island",fontsize=font_size)
plt.ylabel("bill_length_mm",fontsize=font_size)

plt.show()
```

```python
sns.boxplot(x = 'species', y = 'body_mass_g', data = df_penguins, hue = "species",palette=colors)
plt.title("Body mass of each species",fontsize=font_size)
plt.xlabel("Species",fontsize=font_size)
plt.ylabel("Body Mass (g)",fontsize=font_size)
```

```python
#bill length and depth wrt species 

fig,axs = plt.subplots(ncols = 2)
fig.tight_layout()

sns.boxplot(y= 'bill_depth_mm', x = 'species', data = df_penguins, ax= axs[0], hue = "species", palette=colors)
sns.boxplot(y= 'bill_length_mm', x = 'species', data = df_penguins, ax= axs[1],hue = "species", palette=colors)
```

```python
# Plot scatter plot with hue as species
sns.scatterplot(data=df_penguins, x="bill_length_mm", y="body_mass_g", hue="species",palette=colors)

# Set plot title and labels
plt.title("Body mass vs bill length",fontsize=font_size)
plt.xlabel("Bill Length (mm)",fontsize=font_size)
plt.ylabel("Body Mass (g)",fontsize=font_size)

# Show the plot
plt.show()
```

```python
# Plot scatter plot with hue as species
sns.scatterplot(data=df_penguins, x="bill_length_mm", y="bill_depth_mm", hue="species",palette=colors)

# Set plot title and labels
plt.title("Bill length vs bill depth",fontsize=font_size)
plt.xlabel("Bill Length (mm)",fontsize=font_size)
plt.ylabel("Bill Depth (mm)",fontsize=font_size)

# Show the plot
plt.show()
```

```python
# Plot scatter plot with hue as species
sns.scatterplot(data=df_penguins, x="bill_length_mm", y="flipper_length_mm", hue="species",palette=colors)

# Set plot title and labels
plt.title("Flipper length vs Bill length",fontsize=font_size)
plt.xlabel("Bill Length (mm)",fontsize=font_size)
plt.ylabel("Flipper Length (mm)",fontsize=font_size)

# Show the plot
plt.show()
```
